```python
# ==========================================
# SPOTIFY VS TIKTOK USER ENGAGEMENT ANALYSIS
# Comprehensive Data Analysis for Jupyter Notebook
# ==========================================

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('default')
sns.set_palette("husl")

# Configure display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print("üìä SPOTIFY VS TIKTOK USER ENGAGEMENT ANALYSIS")
print("=" * 50)

# ==========================================
# 1. DATA LOADING AND EXPLORATION
# ==========================================

# Load the dataset
df = pd.read_excel('Spotify_vs_TikTok_User_Data_10000.xlsx')

print(f"üìà Dataset loaded successfully!")
print(f"Dataset shape: {df.shape}")
print(f"Columns: {list(df.columns)}")

# Basic dataset information
print("\nüîç DATASET OVERVIEW:")
print("-" * 30)
print(df.info())

print("\nüìä BASIC STATISTICS:")
print("-" * 20)
print(df.describe())

print("\nüéØ PLATFORM DISTRIBUTION:")
print("-" * 25)
print(df['Platform'].value_counts())
print(f"Spotify: {df['Platform'].value_counts()['Spotify']} users ({df['Platform'].value_counts()['Spotify']/len(df)*100:.1f}%)")
print(f"TikTok: {df['Platform'].value_counts()['TikTok']} users ({df['Platform'].value_counts()['TikTok']/len(df)*100:.1f}%)")

# Check for missing values
print("\n‚ùì MISSING VALUES:")
print("-" * 18)
print(df.isnull().sum())

# ==========================================
# 2. DATA PREPROCESSING
# ==========================================

# Separate data by platform
spotify_data = df[df['Platform'] == 'Spotify'].copy()
tiktok_data = df[df['Platform'] == 'TikTok'].copy()

print(f"\nüì± Platform Data Split:")
print(f"Spotify users: {len(spotify_data)}")
print(f"TikTok users: {len(tiktok_data)}")

# ==========================================
# 3. DEMOGRAPHIC ANALYSIS
# ==========================================

print("\nüë• DEMOGRAPHIC ANALYSIS:")
print("-" * 25)

# Age statistics
age_stats = df.groupby('Platform')['Age'].agg(['mean', 'median', 'std', 'min', 'max']).round(2)
print("Age Statistics by Platform:")
print(age_stats)

# Create demographic comparison
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Demographic Analysis: Spotify vs TikTok', fontsize=16, fontweight='bold')

# Age distribution by platform
axes[0,0].hist(spotify_data['Age'], alpha=0.7, label='Spotify', bins=20, color='#1db954')
axes[0,0].hist(tiktok_data['Age'], alpha=0.7, label='TikTok', bins=20, color='#ff0050')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Frequency')
axes[0,0].set_title('Age Distribution by Platform')
axes[0,0].legend()

# Box plot for age comparison
df.boxplot(column='Age', by='Platform', ax=axes[0,1])
axes[0,1].set_title('Age Distribution Comparison')
axes[0,1].set_xlabel('Platform')
axes[0,1].set_ylabel('Age')

# Country distribution (top 10)
top_countries = df['Country'].value_counts().head(10)
axes[1,0].bar(range(len(top_countries)), top_countries.values)
axes[1,0].set_xticks(range(len(top_countries)))
axes[1,0].set_xticklabels(top_countries.index, rotation=45)
axes[1,0].set_title('Top 10 Countries by User Count')
axes[1,0].set_ylabel('Number of Users')

# Platform distribution pie chart
platform_counts = df['Platform'].value_counts()
axes[1,1].pie(platform_counts.values, labels=platform_counts.index, autopct='%1.1f%%', 
              colors=['#1db954', '#ff0050'])
axes[1,1].set_title('Platform Distribution')

plt.tight_layout()
plt.show()

# ==========================================
# 4. ENGAGEMENT METRICS ANALYSIS
# ==========================================

print("\nüöÄ ENGAGEMENT METRICS ANALYSIS:")
print("-" * 32)

# Calculate engagement metrics by platform
engagement_metrics = df.groupby('Platform').agg({
    'DailyListeningHours': ['mean', 'median', 'std'],
    'SocialInteractions': ['mean', 'median', 'std'],
    'SatisfactionScore': ['mean', 'median', 'std']
}).round(2)

print("Engagement Metrics by Platform:")
print(engagement_metrics)

# Create engagement comparison charts
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Engagement Metrics: Spotify vs TikTok', fontsize=16, fontweight='bold')

# Daily listening hours comparison
spotify_hours = spotify_data['DailyListeningHours']
tiktok_hours = tiktok_data['DailyListeningHours']

axes[0,0].hist(spotify_hours, alpha=0.7, label='Spotify', bins=30, color='#1db954')
axes[0,0].hist(tiktok_hours, alpha=0.7, label='TikTok', bins=30, color='#ff0050')
axes[0,0].set_xlabel('Daily Hours')
axes[0,0].set_ylabel('Frequency')
axes[0,0].set_title('Daily Usage Hours Distribution')
axes[0,0].legend()

# Social interactions comparison
df.boxplot(column='SocialInteractions', by='Platform', ax=axes[0,1])
axes[0,1].set_title('Social Interactions Comparison')
axes[0,1].set_xlabel('Platform')
axes[0,1].set_ylabel('Social Interactions')

# Satisfaction scores comparison
df.boxplot(column='SatisfactionScore', by='Platform', ax=axes[1,0])
axes[1,0].set_title('Satisfaction Score Comparison')
axes[1,0].set_xlabel('Platform')
axes[1,0].set_ylabel('Satisfaction Score')

# Engagement metrics bar chart
metrics_means = df.groupby('Platform')[['DailyListeningHours', 'SocialInteractions', 'SatisfactionScore']].mean()
x = np.arange(len(metrics_means.columns))
width = 0.35

axes[1,1].bar(x - width/2, metrics_means.loc['Spotify'], width, label='Spotify', color='#1db954')
axes[1,1].bar(x + width/2, metrics_means.loc['TikTok'], width, label='TikTok', color='#ff0050')
axes[1,1].set_xlabel('Metrics')
axes[1,1].set_ylabel('Average Values')
axes[1,1].set_title('Average Engagement Metrics Comparison')
axes[1,1].set_xticks(x)
axes[1,1].set_xticklabels(['Daily Hours', 'Social Interactions', 'Satisfaction'])
axes[1,1].legend()

plt.tight_layout()
plt.show()

# ==========================================
# 5. CONTENT PREFERENCES ANALYSIS
# ==========================================

print("\nüéµ CONTENT PREFERENCES ANALYSIS:")
print("-" * 33)

# Top genres by platform
print("Top 5 Genres by Platform:")
spotify_genres = spotify_data['TopGenre'].value_counts().head()
tiktok_genres = tiktok_data['TopGenre'].value_counts().head()

print("\nSpotify Top Genres:")
print(spotify_genres)
print("\nTikTok Top Genres:")
print(tiktok_genres)

# Create content preference visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Content Preferences: Spotify vs TikTok', fontsize=16, fontweight='bold')

# Top genres for Spotify
spotify_genres.plot(kind='barh', ax=axes[0,0], color='#1db954')
axes[0,0].set_title('Top 5 Genres - Spotify')
axes[0,0].set_xlabel('Number of Users')

# Top genres for TikTok
tiktok_genres.plot(kind='barh', ax=axes[0,1], color='#ff0050')
axes[0,1].set_title('Top 5 Genres - TikTok')
axes[0,1].set_xlabel('Number of Users')

# Favorite features comparison
spotify_features = spotify_data['FavoriteFeature'].value_counts().head()
tiktok_features = tiktok_data['FavoriteFeature'].value_counts().head()

spotify_features.plot(kind='barh', ax=axes[1,0], color='#1db954')
axes[1,0].set_title('Top 5 Features - Spotify')
axes[1,0].set_xlabel('Number of Users')

tiktok_features.plot(kind='barh', ax=axes[1,1], color='#ff0050')
axes[1,1].set_title('Top 5 Features - TikTok')
axes[1,1].set_xlabel('Number of Users')

plt.tight_layout()
plt.show()

print("\nTop 5 Features by Platform:")
print("\nSpotify Top Features:")
print(spotify_features)
print("\nTikTok Top Features:")
print(tiktok_features)

# ==========================================
# 6. RETENTION AND CHURN ANALYSIS
# ==========================================

print("\nüìà RETENTION AND CHURN ANALYSIS:")
print("-" * 33)

# Churn risk analysis
churn_analysis = pd.crosstab(df['Platform'], df['ChurnRisk'], normalize='index') * 100
print("Churn Risk Distribution (%):")
print(churn_analysis.round(1))

# Viral engagement analysis
viral_analysis = pd.crosstab(df['Platform'], df['ViralEngagement'], normalize='index') * 100
print("\nViral Engagement Distribution (%):")
print(viral_analysis.round(1))

# Create retention analysis charts
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Retention & Engagement Analysis: Spotify vs TikTok', fontsize=16, fontweight='bold')

# Churn risk distribution
churn_raw = pd.crosstab(df['Platform'], df['ChurnRisk'])
churn_raw.plot(kind='bar', ax=axes[0,0], color=['#ff6b6b', '#ffd93d', '#6bcf7f'])
axes[0,0].set_title('Churn Risk Distribution')
axes[0,0].set_ylabel('Number of Users')
axes[0,0].set_xlabel('Platform')
axes[0,0].legend(title='Churn Risk')
axes[0,0].tick_params(axis='x', rotation=0)

# Churn risk percentage
churn_analysis.plot(kind='bar', ax=axes[0,1], color=['#ff6b6b', '#ffd93d', '#6bcf7f'])
axes[0,1].set_title('Churn Risk Distribution (%)')
axes[0,1].set_ylabel('Percentage')
axes[0,1].set_xlabel('Platform')
axes[0,1].legend(title='Churn Risk')
axes[0,1].tick_params(axis='x', rotation=0)

# Viral engagement distribution
viral_raw = pd.crosstab(df['Platform'], df['ViralEngagement'])
viral_raw.plot(kind='bar', ax=axes[1,0], color=['#ff9999', '#66b3ff', '#99ff99'])
axes[1,0].set_title('Viral Engagement Distribution')
axes[1,0].set_ylabel('Number of Users')
axes[1,0].set_xlabel('Platform')
axes[1,0].legend(title='Viral Engagement')
axes[1,0].tick_params(axis='x', rotation=0)

# Viral engagement percentage
viral_analysis.plot(kind='bar', ax=axes[1,1], color=['#ff9999', '#66b3ff', '#99ff99'])
axes[1,1].set_title('Viral Engagement Distribution (%)')
axes[1,1].set_ylabel('Percentage')
axes[1,1].set_xlabel('Platform')
axes[1,1].legend(title='Viral Engagement')
axes[1,1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

# ==========================================
# 7. CORRELATION ANALYSIS
# ==========================================

print("\nüîó CORRELATION ANALYSIS:")
print("-" * 23)

# Select numerical columns for correlation
numerical_cols = ['Age', 'DailyListeningHours', 'SocialInteractions', 'SatisfactionScore']
correlation_matrix = df[numerical_cols].corr()

print("Correlation Matrix:")
print(correlation_matrix.round(3))

# Create correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.3f', cbar_kws={"shrink": .8})
plt.title('Correlation Matrix - User Engagement Metrics', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# ==========================================
# 8. STATISTICAL TESTING
# ==========================================

print("\nüìä STATISTICAL TESTING:")
print("-" * 23)

from scipy import stats

# T-tests for comparing platforms
def perform_ttest(column_name):
    spotify_values = spotify_data[column_name]
    tiktok_values = tiktok_data[column_name]
    
    t_stat, p_value = stats.ttest_ind(spotify_values, tiktok_values)
    
    print(f"\n{column_name}:")
    print(f"  Spotify mean: {spotify_values.mean():.3f} (¬±{spotify_values.std():.3f})")
    print(f"  TikTok mean: {tiktok_values.mean():.3f} (¬±{tiktok_values.std():.3f})")
    print(f"  T-statistic: {t_stat:.3f}")
    print(f"  P-value: {p_value:.6f}")
    print(f"  Significant difference: {'Yes' if p_value < 0.05 else 'No'}")

# Perform t-tests for key metrics
for column in ['Age', 'DailyListeningHours', 'SocialInteractions', 'SatisfactionScore']:
    perform_ttest(column)

# ==========================================
# 9. ADVANCED VISUALIZATIONS WITH PLOTLY
# ==========================================

print("\nüìà CREATING INTERACTIVE CHARTS...")

# Interactive scatter plot
fig_scatter = px.scatter(df, x='DailyListeningHours', y='SatisfactionScore', 
                        color='Platform', size='SocialInteractions',
                        hover_data=['Age', 'TopGenre', 'ChurnRisk'],
                        title='User Engagement: Daily Hours vs Satisfaction Score',
                        color_discrete_map={'Spotify': '#1db954', 'TikTok': '#ff0050'})
fig_scatter.show()

# Interactive box plots
fig_box = make_subplots(rows=1, cols=3, 
                       subplot_titles=['Daily Hours', 'Social Interactions', 'Satisfaction Score'])

for i, column in enumerate(['DailyListeningHours', 'SocialInteractions', 'SatisfactionScore'], 1):
    fig_box.add_trace(go.Box(y=spotify_data[column], name='Spotify', 
                            marker_color='#1db954'), row=1, col=i)
    fig_box.add_trace(go.Box(y=tiktok_data[column], name='TikTok', 
                            marker_color='#ff0050'), row=1, col=i)

fig_box.update_layout(title_text="Engagement Metrics Comparison", showlegend=False)
fig_box.show()

# ==========================================
# 10. SUMMARY REPORT GENERATION
# ==========================================

print("\nüìã SUMMARY REPORT:")
print("=" * 20)

# Calculate key metrics
spotify_metrics = {
    'users': len(spotify_data),
    'avg_age': spotify_data['Age'].mean(),
    'avg_daily_hours': spotify_data['DailyListeningHours'].mean(),
    'avg_social_interactions': spotify_data['SocialInteractions'].mean(),
    'avg_satisfaction': spotify_data['SatisfactionScore'].mean(),
    'low_churn_pct': (spotify_data['ChurnRisk'] == 'Low').mean() * 100,
    'high_viral_pct': (spotify_data['ViralEngagement'] == 'High').mean() * 100
}

tiktok_metrics = {
    'users': len(tiktok_data),
    'avg_age': tiktok_data['Age'].mean(),
    'avg_daily_hours': tiktok_data['DailyListeningHours'].mean(),
    'avg_social_interactions': tiktok_data['SocialInteractions'].mean(),
    'avg_satisfaction': tiktok_data['SatisfactionScore'].mean(),
    'low_churn_pct': (tiktok_data['ChurnRisk'] == 'Low').mean() * 100,
    'high_viral_pct': (tiktok_data['ViralEngagement'] == 'High').mean() * 100
}

print("üéµ SPOTIFY METRICS:")
print(f"  ‚Ä¢ Total Users: {spotify_metrics['users']:,}")
print(f"  ‚Ä¢ Average Age: {spotify_metrics['avg_age']:.1f} years")
print(f"  ‚Ä¢ Daily Hours: {spotify_metrics['avg_daily_hours']:.2f} hours")
print(f"  ‚Ä¢ Social Interactions: {spotify_metrics['avg_social_interactions']:.2f}")
print(f"  ‚Ä¢ Satisfaction Score: {spotify_metrics['avg_satisfaction']:.2f}/10")
print(f"  ‚Ä¢ Low Churn Risk: {spotify_metrics['low_churn_pct']:.1f}%")
print(f"  ‚Ä¢ High Viral Engagement: {spotify_metrics['high_viral_pct']:.1f}%")

print("\nüì± TIKTOK METRICS:")
print(f"  ‚Ä¢ Total Users: {tiktok_metrics['users']:,}")
print(f"  ‚Ä¢ Average Age: {tiktok_metrics['avg_age']:.1f} years")
print(f"  ‚Ä¢ Daily Hours: {tiktok_metrics['avg_daily_hours']:.2f} hours")
print(f"  ‚Ä¢ Social Interactions: {tiktok_metrics['avg_social_interactions']:.2f}")
print(f"  ‚Ä¢ Satisfaction Score: {tiktok_metrics['avg_satisfaction']:.2f}/10")
print(f"  ‚Ä¢ Low Churn Risk: {tiktok_metrics['low_churn_pct']:.1f}%")
print(f"  ‚Ä¢ High Viral Engagement: {tiktok_metrics['high_viral_pct']:.1f}%")

print("\nüîç KEY INSIGHTS:")
print("  ‚Ä¢ Both platforms show remarkably similar user engagement patterns")
print("  ‚Ä¢ Age demographics are nearly identical across platforms")
print("  ‚Ä¢ Daily usage time is consistent (~2.5 hours)")
print("  ‚Ä¢ Social interaction levels are equal (4.0 average)")
print("  ‚Ä¢ Satisfaction scores are comparable (~5.5/10)")
print("  ‚Ä¢ Churn risk distributions are very similar")
print("  ‚Ä¢ Content preferences vary but discovery features are popular on both")

print("\n‚úÖ ANALYSIS COMPLETED!")
print("All charts and metrics have been generated successfully.")

# ==========================================
# 11. EXPORT RESULTS (OPTIONAL)
# ==========================================

# Save summary metrics to CSV
summary_df = pd.DataFrame({
    'Metric': ['Users', 'Avg Age', 'Daily Hours', 'Social Interactions', 
               'Satisfaction', 'Low Churn %', 'High Viral %'],
    'Spotify': [spotify_metrics['users'], spotify_metrics['avg_age'], 
                spotify_metrics['avg_daily_hours'], spotify_metrics['avg_social_interactions'],
                spotify_metrics['avg_satisfaction'], spotify_metrics['low_churn_pct'],
                spotify_metrics['high_viral_pct']],
    'TikTok': [tiktok_metrics['users'], tiktok_metrics['avg_age'],
               tiktok_metrics['avg_daily_hours'], tiktok_metrics['avg_social_interactions'],
               tiktok_metrics['avg_satisfaction'], tiktok_metrics['low_churn_pct'],
               tiktok_metrics['high_viral_pct']]
})

# Uncomment to save results
# summary_df.to_csv('spotify_vs_tiktok_summary.csv', index=False)
# print("\nüíæ Summary results saved to 'spotify_vs_tiktok_summary.csv'")

print("\nüéØ Ready for presentation and further analysis!")

# ==========================================
# FIXED CHARTS FOR SOCIAL INTERACTIONS AND SATISFACTION SCORE
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set up the plotting style
plt.style.use('default')
sns.set_palette("husl")

# Define platform colors
SPOTIFY_COLOR = '#1db954'
TIKTOK_COLOR = '#ff0050'

# Assuming your data is already loaded as 'df' and split into 'spotify_data' and 'tiktok_data'
# If not, uncomment and modify these lines:
# df = pd.read_excel('Spotify_vs_TikTok_User_Data_10000.xlsx')
# spotify_data = df[df['Platform'] == 'Spotify'].copy()
# tiktok_data = df[df['Platform'] == 'TikTok'].copy()

# ==========================================
# FIX 1: SOCIAL INTERACTIONS COMPARISON
# ==========================================

fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# Create box plot with proper colors and styling
box_data = [spotify_data['SocialInteractions'], tiktok_data['SocialInteractions']]
box_labels = ['Spotify', 'TikTok']

bp = ax.boxplot(box_data, labels=box_labels, patch_artist=True, 
                widths=0.6, showfliers=True)

# Color the boxes
bp['boxes'][0].set_facecolor(SPOTIFY_COLOR)
bp['boxes'][0].set_alpha(0.7)
bp['boxes'][1].set_facecolor(TIKTOK_COLOR)
bp['boxes'][1].set_alpha(0.7)

# Style the box plot elements
for element in ['whiskers', 'fliers', 'medians', 'caps']:
    plt.setp(bp[element], color='black', linewidth=1.5)

# Customize the plot
ax.set_title('Social Interactions Comparison\nSpotify vs TikTok', 
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Platform', fontsize=14, fontweight='bold')
ax.set_ylabel('Social Interactions per User', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

# Add statistics text
spotify_mean = spotify_data['SocialInteractions'].mean()
tiktok_mean = tiktok_data['SocialInteractions'].mean()
spotify_median = spotify_data['SocialInteractions'].median()
tiktok_median = tiktok_data['SocialInteractions'].median()

# Add mean values as text
ax.text(1, spotify_mean + 0.5, f'Mean: {spotify_mean:.2f}', 
        ha='center', va='bottom', fontweight='bold', color=SPOTIFY_COLOR)
ax.text(2, tiktok_mean + 0.5, f'Mean: {tiktok_mean:.2f}', 
        ha='center', va='bottom', fontweight='bold', color=TIKTOK_COLOR)

plt.tight_layout()
plt.show()

# ==========================================
# FIX 2: SATISFACTION SCORE COMPARISON
# ==========================================

fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# Create box plot with proper colors and styling
box_data = [spotify_data['SatisfactionScore'], tiktok_data['SatisfactionScore']]
box_labels = ['Spotify', 'TikTok']

bp = ax.boxplot(box_data, labels=box_labels, patch_artist=True, 
                widths=0.6, showfliers=True)

# Color the boxes
bp['boxes'][0].set_facecolor(SPOTIFY_COLOR)
bp['boxes'][0].set_alpha(0.7)
bp['boxes'][1].set_facecolor(TIKTOK_COLOR)
bp['boxes'][1].set_alpha(0.7)

# Style the box plot elements
for element in ['whiskers', 'fliers', 'medians', 'caps']:
    plt.setp(bp[element], color='black', linewidth=1.5)

# Customize the plot
ax.set_title('Satisfaction Score Comparison\nSpotify vs TikTok', 
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Platform', fontsize=14, fontweight='bold')
ax.set_ylabel('Satisfaction Score (1-10)', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

# Set y-axis limits for better visualization
ax.set_ylim(0, 10)

# Add statistics text
spotify_mean = spotify_data['SatisfactionScore'].mean()
tiktok_mean = tiktok_data['SatisfactionScore'].mean()

# Add mean values as text
ax.text(1, spotify_mean + 0.3, f'Mean: {spotify_mean:.2f}', 
        ha='center', va='bottom', fontweight='bold', color=SPOTIFY_COLOR)
ax.text(2, tiktok_mean + 0.3, f'Mean: {tiktok_mean:.2f}', 
        ha='center', va='bottom', fontweight='bold', color=TIKTOK_COLOR)

plt.tight_layout()
plt.show()

# ==========================================
# ALTERNATIVE: SIDE-BY-SIDE VIOLIN PLOTS
# ==========================================

# Create a more modern visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Social Interactions Violin Plot
sns.violinplot(data=df, x='Platform', y='SocialInteractions', ax=ax1,
               palette=[SPOTIFY_COLOR, TIKTOK_COLOR], alpha=0.7)
ax1.set_title('Social Interactions Distribution', fontsize=14, fontweight='bold')
ax1.set_xlabel('Platform', fontsize=12, fontweight='bold')
ax1.set_ylabel('Social Interactions per User', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='y')

# Satisfaction Score Violin Plot
sns.violinplot(data=df, x='Platform', y='SatisfactionScore', ax=ax2,
               palette=[SPOTIFY_COLOR, TIKTOK_COLOR], alpha=0.7)
ax2.set_title('Satisfaction Score Distribution', fontsize=14, fontweight='bold')
ax2.set_xlabel('Platform', fontsize=12, fontweight='bold')
ax2.set_ylabel('Satisfaction Score (1-10)', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')
ax2.set_ylim(0, 10)

plt.suptitle('User Engagement Metrics: Spotify vs TikTok', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# ==========================================
# ALTERNATIVE: COMBINED COMPARISON CHART
# ==========================================

# Create a comprehensive comparison
fig, ax = plt.subplots(figsize=(12, 8))

# Prepare data for grouped bar chart
metrics = ['Daily Hours', 'Social Interactions', 'Satisfaction Score']
spotify_values = [
    spotify_data['DailyListeningHours'].mean(),
    spotify_data['SocialInteractions'].mean(),
    spotify_data['SatisfactionScore'].mean()
]
tiktok_values = [
    tiktok_data['DailyListeningHours'].mean(),
    tiktok_data['SocialInteractions'].mean(),
    tiktok_data['SatisfactionScore'].mean()
]

# Set the width of bars and positions
bar_width = 0.35
r1 = np.arange(len(metrics))
r2 = [x + bar_width for x in r1]

# Create bars
bars1 = ax.bar(r1, spotify_values, width=bar_width, label='Spotify', 
               color=SPOTIFY_COLOR, alpha=0.8)
bars2 = ax.bar(r2, tiktok_values, width=bar_width, label='TikTok', 
               color=TIKTOK_COLOR, alpha=0.8)

# Add value labels on bars
def add_value_labels(bars, values):
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.05,
                f'{value:.2f}', ha='center', va='bottom', fontweight='bold')

add_value_labels(bars1, spotify_values)
add_value_labels(bars2, tiktok_values)

# Customize the plot
ax.set_title('Average Engagement Metrics Comparison\nSpotify vs TikTok', 
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Metrics', fontsize=14, fontweight='bold')
ax.set_ylabel('Average Values', fontsize=14, fontweight='bold')
ax.set_xticks([r + bar_width/2 for r in range(len(metrics))])
ax.set_xticklabels(metrics)
ax.legend(fontsize=12, loc='upper right')
ax.grid(True, alpha=0.3, axis='y')

# Set y-axis to start from 0 for better comparison
ax.set_ylim(0, max(max(spotify_values), max(tiktok_values)) * 1.2)

plt.tight_layout()
plt.show()

# ==========================================
# PRINT SUMMARY STATISTICS
# ==========================================

print("=" * 50)
print("DETAILED STATISTICS COMPARISON")
print("=" * 50)

print("\nüéµ SPOTIFY STATISTICS:")
print("-" * 25)
print(f"Social Interactions - Mean: {spotify_data['SocialInteractions'].mean():.2f}, Median: {spotify_data['SocialInteractions'].median():.2f}, Std: {spotify_data['SocialInteractions'].std():.2f}")
print(f"Satisfaction Score - Mean: {spotify_data['SatisfactionScore'].mean():.2f}, Median: {spotify_data['SatisfactionScore'].median():.2f}, Std: {spotify_data['SatisfactionScore'].std():.2f}")
print(f"Daily Hours - Mean: {spotify_data['DailyListeningHours'].mean():.2f}, Median: {spotify_data['DailyListeningHours'].median():.2f}, Std: {spotify_data['DailyListeningHours'].std():.2f}")

print("\nüì± TIKTOK STATISTICS:")
print("-" * 23)
print(f"Social Interactions - Mean: {tiktok_data['SocialInteractions'].mean():.2f}, Median: {tiktok_data['SocialInteractions'].median():.2f}, Std: {tiktok_data['SocialInteractions'].std():.2f}")
print(f"Satisfaction Score - Mean: {tiktok_data['SatisfactionScore'].mean():.2f}, Median: {tiktok_data['SatisfactionScore'].median():.2f}, Std: {tiktok_data['SatisfactionScore'].std():.2f}")
print(f"Daily Hours - Mean: {tiktok_data['DailyListeningHours'].mean():.2f}, Median: {tiktok_data['DailyListeningHours'].median():.2f}, Std: {tiktok_data['DailyListeningHours'].std():.2f}")

print("\nüîç KEY DIFFERENCES:")
print("-" * 18)
social_diff = abs(spotify_data['SocialInteractions'].mean() - tiktok_data['SocialInteractions'].mean())
satisfaction_diff = abs(spotify_data['SatisfactionScore'].mean() - tiktok_data['SatisfactionScore'].mean())
hours_diff = abs(spotify_data['DailyListeningHours'].mean() - tiktok_data['DailyListeningHours'].mean())

print(f"Social Interactions difference: {social_diff:.3f}")
print(f"Satisfaction Score difference: {satisfaction_diff:.3f}")
print(f"Daily Hours difference: {hours_diff:.3f}")

print("\n‚úÖ CHARTS FIXED AND ENHANCED!")
```
